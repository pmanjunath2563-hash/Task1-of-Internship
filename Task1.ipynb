```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("sales_data.CSV") 
```

```python
print("Shape of Dataset:",df.shape)
```

```python
print("First 5 rows of dataset:",df.head())
```

```python
print("Column names:\n",df.columns)
```

```python
print("Data Types:\n",df.dtypes)
```

```python
print("Missing Values:\n",df.isnull().sum())
```

```python
print("Duplicate rows:\n",df.duplicated().sum())
```

```python
#removing duplicate rows

df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
txt_columns = df.select_dtypes(include='object').columns
for col in txt_columns:
    df[col] = df[col].str.strip().str.lower()
```

```python
df.columns
```

```python
df.Region
```

```python
df['Region'] = df['Region'].replace({'north':'norrthh'})#changoing col name
```

```python
df.Region
```

```python
if 'Sale_Date' in df.columns:
    df['Sale_Date'] = pd.to_datetime(df['Sale_Date'], errors='coerce')
```

```python
df.Sale_Date
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace('','_')
```

```python
df.columns
```

```python
df.describe()
```

```python
print("AFTER CLEANING THE DATASET:\n")
print(df.info())
```

```python
print("Missing values after cleaning:\n",df.isnull().sum())
```

```python
print("Duplicate rows after cleaning:",df.duplicated().sum())
```

```python
#saving the cleaned dataset
df.to_csv("Sales_data_cleaned.csv", index=False)
print("Cleaned dataset saved as 'Sales_data_cleaned.csv'")
```
